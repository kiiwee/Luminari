```python
import requests
from bs4 import BeautifulSoup
import html2text

def search(query):
    url = "http://localhost:8080/search"
    params = {
        'q': query,
        'format': 'json',

    }
    response = requests.get(url, params=params,cookies={'lang;':'en'})
    return response.json()


def get_text_from_url(results):
    res_l = []
    for result in results:

        url = result.get('url')
        print(url)
        if 'medium.com' in url:
            url = 'https://freedium.cfd/'+url
        html = requests.get(url).text
        soup = BeautifulSoup(html, 'html.parser')
        text = [p.text for p in soup.find_all(['p', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6'])]


        
        ai_res = {}
        ai_res['url'] = url
        ai_res['title'] = result.get('title')
        ai_res['content'] = result.get('content')
        ai_res['text'] = text

        res_l.append(ai_res)

    return res_l


result = search("langchain ollama")
get_text_from_url(result.get('results')[:5])
```
